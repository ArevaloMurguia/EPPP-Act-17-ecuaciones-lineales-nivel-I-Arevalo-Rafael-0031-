#### 🔹 Problema 1: x + 8 = 15

🧠 Procedimiento

x + 8 = 15

x = 15 - 8
x = 7

👉 Resultado: x = 7



```python
# Problema 1: x + 8 = 15

resultado = int(input("Ingrese el resultado: "))
numero_suma = int(input("Ingrese el número que se suma: "))

# Despeje: x = resultado - numero_suma
x = resultado - numero_suma

print(f"El valor de x es: {x}")
```

####🔹 Problema 3: 3x + 4x = 21

🧠 Procedimiento

3x + 4x = 21

7x = 21
x = 21 / 7
x = 3

👉 Resultado: x = 3

```python
# Problema 3: 3x + 4x = 21

resultado = int(input("Ingrese el resultado: "))
coef1 = int(input("Ingrese el primer coeficiente: "))
coef2 = int(input("Ingrese el segundo coeficiente: "))

# Despeje: x = resultado / (coef1 + coef2)
x = resultado / (coef1 + coef2)

print(f"El valor de x es: {x}")
```

#### 🔹 Problema 5: 6a - 2a + a = 20

🧠 Procedimiento

6a - 2a + a = 20

5a = 20
a = 20 / 5
a = 4

👉 Resultado: a = 4

```python
# Problema 5: 6a - 2a + a = 20

resultado = int(input("Ingrese el resultado: "))
coef1 = int(input("Ingrese el primer coeficiente: "))
coef2 = int(input("Ingrese el segundo coeficiente (negativo): "))
coef3 = int(input("Ingrese el tercer coeficiente: "))

# Despeje: a = resultado / (coef1 + coef2 + coef3)
a = resultado / (coef1 + coef2 + coef3)

print(f"El valor de a es: {a}")
```

#### 🔹 Problema 7: 4x + 3 = 2x + 11
#### 🧠 Procedimiento

4x + 3 = 2x + 11

#### Pasamos términos semejantes:

* 4x - 2x = 11 - 3

* 2x = 8

* x = 8 / 2

* x = 4

```python
# Problema 7: 4x + 3 = 2x + 11

lado_izq_x = int(input("Coeficiente de x en lado izquierdo: "))
lado_izq_num = int(input("Número independiente izquierdo: "))
lado_der_x = int(input("Coeficiente de x en lado derecho: "))
lado_der_num = int(input("Número independiente derecho: "))

# Paso 1: 4x - 2x
coef_resultante = lado_izq_x - lado_der_x

# Paso 2: 11 - 3
num_resultante = lado_der_num - lado_izq_num

# Paso 3: x = num_resultante / coef_resultante
x = num_resultante / coef_resultante

print(f"El valor de x es: {x}")
```

#### 🔹 Problema 9: 9m + 4 = 5m + 20

🧠 Procedimiento

9m + 4 = 5m + 20

9m - 5m = 20 - 4
4m = 16
m = 16 / 4
m = 4

👉 Resultado: m = 4

```python
# Problema 9: 9m + 4 = 5m + 20

lado_izq_m = int(input("Coeficiente de m izquierdo: "))
lado_izq_num = int(input("Número independiente izquierdo: "))
lado_der_m = int(input("Coeficiente de m derecho: "))
lado_der_num = int(input("Número independiente derecho: "))

# Despeje: m = (lado_der_num - lado_izq_num) / (lado_izq_m - lado_der_m)
m = (lado_der_num - lado_izq_num) / (lado_izq_m - lado_der_m)

print(f"El valor de m es: {m}")
```

#### 🔹 Problema 11: 6a + 5 = 2a + 21

🧠 Procedimiento

6a + 5 = 2a + 21

6a - 2a = 21 - 5
4a = 16
a = 16 / 4
a = 4

👉 Resultado: a = 4

```python
# Problema 11: 6a + 5 = 2a + 21

lado_izq_a = int(input("Coeficiente de a izquierdo: "))
lado_izq_num = int(input("Número independiente izquierdo: "))
lado_der_a = int(input("Coeficiente de a derecho: "))
lado_der_num = int(input("Número independiente derecho: "))

# Despeje: a = (lado_der_num - lado_izq_num) / (lado_izq_a - lado_der_a)
a = (lado_der_num - lado_izq_num) / (lado_izq_a - lado_der_a)

print(f"El valor de a es: {a}")
```
